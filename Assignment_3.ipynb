# Data Split

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20, random_state=1)


print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier()

# Train the classifier using your training data
rf_classifier.fit(x_train_reshaped, y_train)

# Make predictions on the test set
rf_predictions = rf_classifier.predict(x_test_reshaped)
```

# Support Vector Machine

```python
from sklearn.svm import SVC

# Create an SVM classifier
svm_classifier = SVC()

# Train the classifier using your training data
svm_classifier.fit(x_train_reshaped, y_train)

# Make predictions on the test set
svm_predictions = svm_classifier.predict(x_test_reshaped)
```

# XG Boost

```python

import xgboost as xgb

# Create an XGBoost classifier
xgb_classifier = xgb.XGBClassifier()

# Train the classifier using your training data
xgb_classifier.fit(x_train_reshaped, y_train)

# Make predictions on the test set
xgb_predictions = xgb_classifier.predict(x_test_reshaped)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression classifier
lr_classifier = LogisticRegression()

# Train the classifier using your training data
lr_classifier.fit(x_train_reshaped, y_train)

# Make predictions on the test set
lr_predictions = lr_classifier.predict(x_test_reshaped)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Create a Naive Bayes classifier
nb_classifier = GaussianNB()

# Train the classifier using your training data
nb_classifier.fit(x_train_reshaped, y_train)

# Make predictions on the test set
nb_predictions = nb_classifier.predict(x_test_reshaped)
```

# Performance

```python
from sklearn.metrics import classification_report

# Calculate and print the classification report for each model
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))

print("\nSVM Classification Report:")
print(classification_report(y_test, svm_predictions))

print("\nXGBoost Classification Report:")
print(classification_report(y_test, xgb_predictions))

print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, lr_predictions))

print("\nNaive Bayes Classification Report:")
print(classification_report(y_test, nb_predictions))
```

Percision: It represents the accuracy of positive predictions and indicates how many of the instances classified as positive are actually relevant to the class.

recall:It represents the ability of the classifier to correctly identify positive instances and captures the proportion of relevant instances that are successfully retrieved.

F1-score:It is useful when you want to find a balance between precision and recall, especially when the class distribution is imbalanced.




As we can see XGBoost and Random Forrest had the highest percision, recall, and f1-score where the model can predict the topic each other will use.So, both model have the best performance overall.
